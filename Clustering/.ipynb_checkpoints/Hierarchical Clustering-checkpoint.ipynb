```python
#Import libarary

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.cluster import AgglomerativeClustering 
from sklearn.metrics import silhouette_score 
import scipy.cluster.hierarchy as shc 
from scipy.cluster.hierarchy import linkage 
```

```python
#read file
crime_data = pd.read_csv('https://raw.githubusercontent.com/slmsshk/DataSet/main/crime.csv') 
```

```python
crime_data.head()
```

```python
# 
X = crime_data.iloc[:,2:]
Y=crime_data.iloc[:,:2]
```

```python
z = linkage(X, method="centroid",metric="euclidean")

plt.figure(figsize=(15, 5));
plt.title('Hierarchical Clustering Dendrogram');
plt.xlabel('State',color='blue');
shc.dendrogram(
    z,
    leaf_rotation=0.,  # rotates the x axis labels
    leaf_font_size=8.,  # font size for the x axis labels
    labels=crime_data.abbr.values
)
plt.show()
```

```python
#build model using 3 cluster
h_complete=AgglomerativeClustering(n_clusters=3,linkage='ward',affinity = "euclidean").fit(X) 
```

```python
#view cluster assigned number
cluster_labels=pd.Series(h_complete.labels_,name='Cluster')
silhouette_score(X,cluster_labels),cluster_labels.head(),
```

```python
group = pd.concat((Y,cluster_labels),axis=1).sort_values(by='Cluster')
group[group['Cluster']==1]
```
