# Wine DataSet.

```python
# import wine DataSet
import sklearn.datasets
```

```python
target_names=sklearn.datasets.load_wine().target_names
feature_names =sklearn.datasets.load_wine().feature_names
```

```python
X=sklearn.datasets.load_wine().data
Y=sklearn.datasets.load_wine().target
X.shape
```

# DTC

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
plt.figure(figsize=(15,10),dpi=512)
plot_tree(dtc,class_names=target_names,feature_names=feature_names,filled=True)
plt.show()
```

# Stacking

```python
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
```

```python
# create the sub models
estimators = []
model1 = LogisticRegression()
estimators.append(('logistic', model1))
model2 = DecisionTreeClassifier()
estimators.append(('cart', model2))
model3 = SVC()
estimators.append(('svm', model3))

# create the ensemble model
ensemble = VotingClassifier(estimators)
ensemble.fit(x_train,y_train)
ensemble.score(x_test,y_test)
```

# LGBM

```python
import lightgbm as lgb

d_train = lgb.Dataset(x_train, label=y_train)

params = {}
params['learning_rate'] = 0.002, # 0.1,0.3
params['boosting_type'] = 'gbdt' # dart
params['objective'] = 'multiclass'
params['metric'] = 'softmax'
params['sub_feature'] = 0.5 
params['num_leaves'] = 10
params['min_data'] = 5
params['max_depth'] = 10
params['num_class']=3

lgb_=lgb
clf=lgb_.train(params,d_train)
```
